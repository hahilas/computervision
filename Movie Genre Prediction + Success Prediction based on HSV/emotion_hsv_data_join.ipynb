```python
def concatenate_datasets(hsv_data, emotion_data):
  emotion_newdata = emotion_data.copy()
  emotion_newdata = emotion_newdata[['imdbId', 'Angry', 'Sad', 'Neutral', 'Disgust', 'Surprise', 'Fear', 'Happy']]
  emotion_hsv_data = pd.merge(hsv_data,emotion_newdata,on='imdbId')
  emotion_hsv_data.drop(emotion_hsv_data.iloc[:, 0:2], inplace = True, axis = 1)
  return emotion_hsv_data
```

```python
combined_thriller = concatenate_datasets(thriller, thriller_emotions)
combined_scifi = concatenate_datasets(scifi, scifi_emotions)
combined_romance = concatenate_datasets(romance, romance_emotions)
combined_horror = concatenate_datasets(horror, horror_emotions)
combined_drama = concatenate_datasets(drama, drama_emotions)
combined_crime = concatenate_datasets(crime, crime_emotions)
combined_comedy = concatenate_datasets(comedy, comedy_emotions)
combined_animation = concatenate_datasets(animation, animation_emotions)
combined_adventure = concatenate_datasets(adventure, adventure_emotions)
combined_action = concatenate_datasets(action, action_emotions)
```

```python
import inspect
def retrieve_name(var):
    callers_local_vars = inspect.currentframe().f_back.f_locals.items()
    return [var_name for var_name, var_val in callers_local_vars if var_val is var]
```

```python
combined_datasets = [combined_thriller, combined_scifi, combined_romance, combined_horror, combined_drama, combined_crime, combined_comedy, combined_animation, combined_adventure, combined_action]
for data in combined_datasets:
  print(retrieve_name(data)[0],':\n', data[data.isnull().any(axis=1)].shape[0])
```
