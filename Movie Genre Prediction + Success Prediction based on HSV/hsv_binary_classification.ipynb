```python
thriller = pd.read_csv('./Dataset/DATA_AFTER_HSV_EXTRACTION/THRILLER_HSV.csv')
scifi = pd.read_csv('./Dataset/DATA_AFTER_HSV_EXTRACTION/SCIFI_HSV.csv')
romance = pd.read_csv('./Dataset/DATA_AFTER_HSV_EXTRACTION/ROMANCE_HSV.csv')
horror = pd.read_csv('./Dataset/DATA_AFTER_HSV_EXTRACTION/HORROR_HSV.csv')
drama = pd.read_csv('./Dataset/DATA_AFTER_HSV_EXTRACTION/DRAMA_HSV.csv')
crime = pd.read_csv('./Dataset/DATA_AFTER_HSV_EXTRACTION/CRIME_HSV.csv')
comedy = pd.read_csv('./Dataset/DATA_AFTER_HSV_EXTRACTION/COMEDY_HSV.csv')
animation = pd.read_csv('./Dataset/DATA_AFTER_HSV_EXTRACTION/ANIMATION_HSV.csv')
adventure = pd.read_csv('./Dataset/DATA_AFTER_HSV_EXTRACTION/ADVENTURE_HSV.csv')
action = pd.read_csv('./Dataset/DATA_AFTER_HSV_EXTRACTION/ACTION_HSV.csv')
```

```python
datasets = [thriller, scifi, romance, horror, drama, crime, comedy, animation, adventure, action]
```

```python
threshold = 1.5
```

```python
for data in datasets:
  data.loc[(data['TMDB Revenue/Budget'] < threshold), 'Successful?'] = 0
  data.loc[(data['TMDB Revenue/Budget'] >= threshold), 'Successful?'] = 1
  data["Successful?"] = data["Successful?"].astype(int)
```

```python
import inspect
def retrieve_name(var):
    callers_local_vars = inspect.currentframe().f_back.f_locals.items()
    return [var_name for var_name, var_val in callers_local_vars if var_val is var]
```

```python
for data in datasets:
  print(retrieve_name(data)[0], '\n', 'Not successful: ', data[data['Successful?'] == 0].shape[0], 'Successful: ', data[data['Successful?'] == 1].shape[0])
```

```python
from sklearn.model_selection import train_test_split # splitting the data
from sklearn.linear_model import LogisticRegression # model algorithm
from sklearn.preprocessing import StandardScaler # data normalization
import warnings
```

```python
def data_selection(data):
  newdata = data.copy()
  newdata = newdata[['h','s','v', 'Successful?']]
  return newdata
```

```python
def train_test_split(data):
  lr_data = data_selection(data)
  train_data = lr_data.sample(frac=0.8, random_state=0)
  test_data = lr_data.drop(train_data.index)
  return train_data, test_data
```

```python
scaler = StandardScaler()

def variable_marker(data):
  hsv_df = data[['h', 's', 'v']]
  scaled_X = scaler.fit_transform(hsv_df)
  X = pd.DataFrame(scaled_X, columns = hsv_df.columns, index = hsv_df.index)
  y = data[['Successful?']]
  return X,y
```

```python
 import sklearn
from sklearn.model_selection import train_test_split # splitting the data
from sklearn.linear_model import LogisticRegression # model algorithm
from sklearn.metrics import accuracy_score, f1_score
def logistic_reg(data):  
  train, test = train_test_split(data)
  train_X, train_y = variable_marker(train)
  test_X, test_y = variable_marker(test)

  lr = LogisticRegression(C = 0.1, solver = 'liblinear')
  lr.fit(train_X,train_y)
  y_pred = lr.predict(test_X)

  # predictions = model.predict(test_X)
  sklearn.metrics.f1_score
  
  return accuracy_score(test_y, y_pred), f1_score(test_y, y_pred)
```

```python
warnings.filterwarnings("ignore")
for data in datasets:
  print(retrieve_name(data)[0],'\n', 'accuracy score: ', logistic_reg(data)[0], 'f1 score: ', logistic_reg(data)[1])
```
